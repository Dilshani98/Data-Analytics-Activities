```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Creating actual and predicted labels

# Assumed values
actual = (
    ['Healthy']*15 +
    ['Sick']*15
)

# Predicted values
predicted = (
    ['Healthy']*14 + ['Sick']*1 +     # Healthy predictions
    ['Healthy']*2 + ['Sick']*13       # Sick predictions
)

# Generate confusion matrix
cm = confusion_matrix(
    actual,
    predicted,
    labels=['Healthy','Sick']
)

# Plot confusion matrix
plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt='d',
    cmap='Blues',
    xticklabels=['Healthy','Sick'],
    yticklabels=['Healthy','Sick']
)

plt.xlabel("Predicted Class")
plt.ylabel("Actual Class")
plt.title("Confusion Matrix for Healthcare Classification Model")

plt.show()
```

According to the below structure

Columns → Predicted Positive/Negative
Rows → Actual Positive/Negative
Cells → TP, FP, FN, TN

```python
import numpy as np
import matplotlib.pyplot as plt

# Confusion matrix values
cm = np.array([
    [13,2],   # TP, FN
    [1,14]    # FP, TN
])

fig, ax = plt.subplots(figsize=(8,6))

# Create matrix
ax.matshow(cm)

# Labels inside boxes
labels = [
    ['TP\n13','FN\n2'],
    ['FP\n1','TN\n14']
]

for i in range(2):
    for j in range(2):
        ax.text(
            j,
            i,
            labels[i][j],
            ha='center',
            va='center',
            fontsize=14
        )

# Axis labels
ax.set_xticks([0,1])
ax.set_yticks([0,1])

ax.set_xticklabels([
    'Predicted Sick (+)',
    'Predicted Healthy (-)'
])

ax.set_yticklabels([
    'Actual Sick (+)',
    'Actual Healthy (-)'
])

plt.title("Healthcare System Confusion Matrix")
plt.show()
```

The test dataset consisted of 30 patient records.

The model misclassified three records, including two False Negatives and one False Positive.

Assumption is 15 sick patients and 15 healthy

Based on the data, the model identified 13 Sick patients and 14 Healthy patients, while incorrectly classifying two Sick patients as Healthy and one Healthy patient as Sick.
